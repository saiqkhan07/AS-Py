```python
Math = int(input("Enter the Marks of math = "));
ISL = int(input("Enter the Marks of ISL = "));
ENG = int(input("Enter the Marks of ENG = "));
BIO = int(input("Enter the Marks of BIO = "));
CS = int(input("Enter the Marks of CS = "));
TotalMarks = 500
percent = ((Math+ISL+ENG+BIO+CS)/TotalMarks) * 100;
print(percent)
if percent <= 100 and percent >= 80:
    print("Grade A+");
elif percent < 80 and percent >= 70:
    print("Percentage = ",percent, "Grade A");
elif percent < 70 and percent >= 60:
    print("Grade B");
elif percent < 60 and percent >= 50:
    print("Grade C");
elif percent < 50 and percent >= 40:
    print("Grade D");
elif percent < 40 and percent >= 33:
    print("Grade E");
elif percent < 33 and percent >= 0:
    print("Fail");
elif percent < 0 or percent > 100:
    print("Enter from 0 - 100");
```
