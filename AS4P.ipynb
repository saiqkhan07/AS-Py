```python
def addition (a, b):  
    return a + b
def subtraction (a, b):  
    return a - b
def multiplication (a, b):  
    return a * b
def division (a, b):  
    return a / b
def power (a, b):  
    return a**b


print ("1. Addition")
print ("2. Subtraction")
print ("3. Multiplication")
print ("4. Division")
print ("5. Power")
operation = int(input("Enter your choice from (1 - 5): "))
a = int(input("Enter number 1 : "))
b = int(input("Enter number 2 : "))
if operation == 1:
   print(a, " + ", b, " = ", addition (a, b))
elif operation == 2:
   print(a, " - ", b, " = ", subtraction (a, b))
elif operation == 3:
   print(a, " * ", b, " = ", multiplication (a, b))
elif operation == 4:
   print(a, " / ", b, " = ", division (a, b))
elif operation == 5:
   print(a, " power ", b, " = ", power (a, b))
else:
    print("Error")
```
